```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
```

```python
# Read Data
data = pd.read_csv('full_sensor_data--only_04-08-17.csv', usecols=['datetime','temperature', 'humidity', 'co2'])
data.describe()
```

```python
# Eliminating outliers

data = data [data.temperature > 0]
data. describe()
```

```python
# Identifying correlation between variables 

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.size']=11
plt.rcParams['figure.figsize']=[15.0, 5.5]
fig, axes=plt.subplots(nrows=1, ncols=2)
data.plot(kind='scatter', x='humidity', y='co2', ax=axes[0], color='Blue');
data.plot(kind='scatter', x='humidity', y='temperature', ax=axes[1], color='Red');
plt.tight_layout()    
```

```python
 # Scatter Matrix of the Variables
from pandas.tools.plotting import scatter_matrix

grouped = data.groupby(level=0)
dfgroup = grouped.mean()

scatter_matrix(dfgroup,figsize=(12, 12), diagonal='kde')
```
