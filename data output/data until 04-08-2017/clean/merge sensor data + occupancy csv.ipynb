```python
# Load occupance data as a dataframe with the 'datetime' column as its index
peeps = 'occupancy_data.csv'
df = pd.read_csv(peeps, index_col='datetime', parse_dates=True)
# add new datetime column "dt" which is a copy of datetime
df['dt'] = df.index
```

```python
print(df.head())
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df
```

```python
# create a new column truncated to minute precision
df['dtm'] = df['dt'].values.astype('<M8[m]')
df
```

```python
# show only datetime truncated by minute and count_total 
df1 = df[['dtm','count_total']]
df1
```

```python
# drop duplicates, keeping the row with the highest value
df1 = df1.groupby('dtm', group_keys=False).apply(lambda x: x.ix[x.count_total.idxmax()])
df1
```

```python
# drop dtm column
del df1['dtm']
# save df1 to csv
df1.to_csv('count_total.csv', sep=',')
df1
```

```python
#Load sensor data as a dataframe with the 'datetime' column as its index
sensor = 'sensor_data.csv'
sensor_data = pd.read_csv(sensor, index_col='datetime', parse_dates=True)
sensor_data['dt'] = sensor_data.index
```

```python
print(sensor_data.head())
```

```python
sensor_data.describe()
```

```python
# create a new column truncated to minute precision
sensor_data['dtm'] = sensor_data['dt'].values.astype('<M8[m]')
```

```python
# show only datetime truncated by minute and count_total of people
sensor_data1 = sensor_data[['dtm','temperature','humidity','co2','light','noise','bluetooth_devices','bluetooth_non_personal_devices']]
```

```python
# drop duplicates, keeping the row with the highest value of temp
sensor_data1 = sensor_data1.groupby('dtm', group_keys=False).apply(lambda x: x.ix[x.temperature.idxmax()])
sensor_data1
```

```python
# drop dtm column
del sensor_data1['dtm']
# save sensor_data1 to csv
sensor_data1.to_csv('sensor_data1.csv', sep=',')
```

```python
#merge newly saved csv files
csv1 = pd.read_csv('sensor_data1.csv')
csv2 = pd.read_csv('count_total.csv')
merged = csv1.merge(csv2, on="dtm", how="outer").fillna(method='ffill')
merged.to_csv("sensor_data1_count_total.csv", index=False)
```
