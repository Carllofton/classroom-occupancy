```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
output1 = pd.read_csv('output1.csv')
output1.head()
```

```python
output1.describe()
```

```python
output1 = output1[['datetime','temperature','humidity','co2','light','noise','bluetooth_devices','count_total']]
output1.head()
```

```python
output1 =output1.dropna()
output1.head()
```

```python
#correlation on output1 data
corrmat1 = output1.corr()
corrmat1
```

```python
#a graph to entertain the author
sns.heatmap(corrmat1, vmax=.8, square=True)
plt.show()
```

```python
#multiple regression using count_total as y
x = output1[['temperature','humidity','co2','light','noise','bluetooth_devices']]
y = output1['count_total']

## fit a OLS model 
x = sm.add_constant(x)
result = smf.ols('count_total ~ temperature + humidity + co2 + light + noise + bluetooth_devices', data=output1).fit()
result.summary()
```
